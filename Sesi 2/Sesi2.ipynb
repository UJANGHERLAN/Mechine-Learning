```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
kota = ['Kota 1', 'Kota 2', 'Kota 3', 'Kota 4', 'Kota 5', 'Kota 6', 'Kota 7', 'Kota 8']
pendapatan = np.array([5, 10, 20, 8, 4, 6, 12, 15])  # Pendapatan rata-rata (dalam ribuan $)
penjualan = np.array([27, 46, 73, 40, 30, 28, 46, 59])  # Penjualan pizza (dalam ribuan buah)
```

```python
# 1. Scatter Diagram
plt.figure(figsize=(8, 6))
plt.scatter(pendapatan, penjualan, color='blue', label='Data')

# Menambahkan label untuk setiap titik
for i in range(len(kota)):
    plt.text(pendapatan[i], penjualan[i], kota[i])

# Regresi Linear
slope, intercept = np.polyfit(pendapatan, penjualan, 1)
plt.plot(pendapatan, slope * pendapatan + intercept, color='red', label='Regresi Linear')

plt.title('Scatter Diagram Pendapatan vs Penjualan Pizza')
plt.xlabel('Pendapatan Rata-rata (ribuan $)')
plt.ylabel('Penjualan Pizza (ribuan buah)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 2. Kedua variabel berkolerasi?
# Untuk mengetahui apakah kedua variabel berkolerasi, kita dapat menggunakan koefisien korelasi Pearson
correlation = np.corrcoef(pendapatan, penjualan)[0, 1]
if correlation > 0:
    print("Kedua variabel tersebut berkolerasi positif.")
elif correlation < 0:
    print("Kedua variabel tersebut berkolerasi negatif.")
else:
    print("Tidak ada korelasi antara kedua variabel tersebut.")
```

```python
# 3. Menghitung slope, intercept, SSR, SSE, STT, R
# Menghitung slope dan intercept menggunakan metode least squares regression
slope, intercept = np.polyfit(pendapatan, penjualan, 1)

# Menghitung SSR, SSE, dan STT
predicted_penjualan = slope * pendapatan + intercept
SSR = np.sum((predicted_penjualan - np.mean(penjualan)) ** 2)
SSE = np.sum((penjualan - predicted_penjualan) ** 2)
STT = SSR + SSE

# Menghitung koefisien determinasi (R)
R = SSR / STT

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"SSR: {SSR}")
print(f"SSE: {SSE}")
print(f"STT: {STT}")
print(f"Koefisien Determinasi (R): {R}")
```
