```python
import matplotlib.pyplot as plt 
import numpy as np 
from scipy.stats import linregress
```

```python
pendapatanRatarata = np.array([5,10,20,8,4,6,12,15])
penjualanPizza = np.array([27,46,73,40,30,28,46,59])
```

```python
plt.scatter(pendapatanRatarata, penjualanPizza, )
plt.title('Scatter Diagram Pendapatan Ratarata vs Penjualan Pizza')
plt.xlabel('Pendapatan Ratarata(1000$)')
plt.ylabel('Penjualan Pizza(1000 Buah)')
plt.show()
```

```python
correlation_coefficient = np.corrcoef(pendapatanRatarata, penjualanPizza)[0, 1]
print(f"Korelasi antara pendapatan rata-rata dan penjualan pizza: {correlation_coefficient}")
```

```python
slope, intercept, r_value, p_value, std_err = linregress(pendapatanRatarata, penjualanPizza)
predicted_values = slope * pendapatanRatarata + intercept
SSR = np.sum((predicted_values - np.mean(penjualanPizza))**2)
SSE = np.sum((penjualanPizza - predicted_values)**2)
SST = SSR + SSE
R_squared = SSR / SST
```

```python
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"SSR: {SSR}")
print(f"SSE: {SSE}")
print(f"SST: {SST}")
print(f"R-squared: {R_squared}")
```
